#### Today, top 10 stations in the city for starting a journey

```python
# Query today's date from the dataframe. Closest date available since last upload by Citi Bike is June 30, 2020.
d='2020-06-30' 
today_date=pd.to_datetime(d).date()

df202006["date"] = pd.to_datetime(df202006["starttime"])
df202006["date"] = df202006["date"].dt.date
df_queried_date = df202006.loc[df202006["date"] == today_date]
```

```python
# Find top stations
df_top_stations = df_queried_date.groupby(["start station id"])\
    ["start station id"].count().sort_values(ascending=False)[0:10].to_frame()
df_top_stations = df_top_stations.rename(columns={"start station id":"number of start stations"})
df_top_stations["start station id"] = df_top_stations.index
df_top_stations.reset_index(drop=True, inplace=True)
# df_top_stations
```

```python
stations = df202006[["start station id", "start station name"]].drop_duplicates()
df_top_stations.merge(stations, how="inner", on="start station id")
```

#### Today, top 10 stations in the city for ending a journey

```python
# Find top stations
df_top_stations = df_queried_date.groupby(["end station id"])\
    ["end station id"].count().sort_values(ascending=False)[0:10].to_frame()
df_top_stations = df_top_stations.rename(columns={"end station id":"number of end stations"})
df_top_stations["end station id"] = df_top_stations.index
df_top_stations.reset_index(drop=True, inplace=True)
# df_top_stations
```

```python
stations = df202006[["end station id", "end station name"]].drop_duplicates()
df_top_stations.merge(stations, how="inner", on="end station id")
```

```python
# Find bottom stations
df_bottom_stations = df_queried_date.groupby(["start station id"])\
    ["start station id"].count().sort_values(ascending=True)[0:10].to_frame()
df_bottom_stations = df_bottom_stations.rename(columns={"start station id":"number of start stations"})
df_bottom_stations["start station id"] = df_bottom_stations.index
df_bottom_stations.reset_index(drop=True, inplace=True)
# df_bottom_stations
```

```python
stations = df202006[["start station id", "start station name"]].drop_duplicates()
df_bottom_stations.merge(stations, how="inner", on="start station id")
```

#### Today, bottom 10 stations in the city for ending a journey

```python
# Find bottom stations
df_bottom_stations = df_queried_date.groupby(["end station id"])\
    ["end station id"].count().sort_values(ascending=True)[0:10].to_frame()
df_bottom_stations = df_bottom_stations.rename(columns={"end station id":"number of end stations"})
df_bottom_stations["end station id"] = df_bottom_stations.index
df_bottom_stations.reset_index(drop=True, inplace=True)
# df_bottom_stations
```
